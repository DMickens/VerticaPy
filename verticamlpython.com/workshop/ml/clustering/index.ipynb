# Clustering

Clustering Algorithms are used to segment the data or to find anomalies. They are most of the time really sensible to un-normalized data. For example if we consider the Titanic dataset, the features 'fare' and 'age' are not living in the same interval. 'fare' can be much higher than the 'age' which can lead to unexpected clusters. 

Let's look on how to build a clustering model in Vertica ML Python.

First, let's import the KMeans algorithm.

```python
from vertica_ml_python.learn.cluster import KMeans
```

We can create a model object. As Vertica has its own model management system, we need to choose a model name with the other parameters.

```python
model = KMeans("KMeans_sm", n_cluster = 6)
```

We can fit the model using the corresponding data.

```python
model.fit("sm_meters", ["latitude", "longitude"])
model.plot()
```

There is no real metric to evaluate unsupervised models. However metrics used during the model computation can help us to understand if the model is good. For example, we generally consider a KMeans model as good when the number of clusters is low and the KMeans score = Between-Cluster SS / Total SS is close to 1.

```python
model.metrics
```
