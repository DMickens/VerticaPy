# Regression

Regressions are specific ML algorithms which are used to predict numerical response columns. For example predicting the salaries of a company employees using their age or predicting the number of attacks a website could face through time.

Most of the Regression Models are sensible to un-normalized data. The most popular Regression Algorithm is the Linear Regression. We always need to verify all the algorithms assumptions before using them.

For example, to create a good Linear Regression Model we need to verify the Gauss Markov assumptions.
 - <b>Linearity</b> : the parameters we are estimating using the OLS method must be themselves linear.
 - <b>Non-Collinearity</b> : the regressors being calculated aren’t perfectly correlated with each other.
 - <b>Exogeneity</b> : the regressors aren’t correlated with the error term.
 - <b>Homoscedasticity</b> : no matter what the values of our regressors might be, the error of the variance is constant.

That's why most of the time when we decided to choose this type of models, we need to normalize and decompose the data (other models like Random Forest are robust to un-normalized and correlated data). If we don't follow the assumptions, it is possible to get unexpected results (example: negative R2)

Let's predict the Total Charges of the telco customers using their respective tenure. We can import the model we need.

```python
from vertica_ml_python.learn.linear_model import LinearRegression
```

We can create a model object. As Vertica has its own model management system, we need to choose a model name with the other parameters.

```python
model = LinearRegression("LR_churn")
```

We can fit the model using the corresponding data.

```python
model.fit("churn", ["tenure"], "TotalCharges")
model.plot()
```

To evaluate the model, we can use different metrics.

```python
model.regression_report()
```

We did not split the data into train and test which will be more relevant. The purpose is to understand all the possible metrics to evaluate a Regression. The most famous one is the R2 for which the closer it is to 1, the better it is (most of the times).
