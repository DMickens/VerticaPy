# vDataFrame[].plot

```python
vDataFrame[].plot(ts: str,
                  columns: list = [], 
                  by: str = "",
                  start_date: str = "",
                  end_date: str = "",
                  color: str = '#214579',
                  area: bool = False)
```

Draws a time series of the vcolumn.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td><div class="param_name">ts</div></td> <td><div class="type">str</div></td> <td><div class = "no">&#10060;</div></td> <td>TS (Time Series) vcolumn used to order the data. The vcolumn type must be date like (date, datetime, timestamp...) or numerical.</td> </tr>
    <tr> <td><div class="param_name">by</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>vcolumn used to partition the TS.</td> </tr>
    <tr> <td><div class="param_name">start_date</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Input Start Date. For example, time = '03-11-1993' will filter the data when 'ts' is lesser than November 1993 the 3rd.</td> </tr>
    <tr> <td><div class="param_name">end_date</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Input End Date. For example, time = '03-11-1993' will filter the data when 'ts' is greater than November 1993 the 3rd.</td> </tr>
    <tr> <td><div class="param_name">color</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Color of the TS.</td> </tr>
    <tr> <td><div class="param_name">area</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to True, draw an Area Plot.</td> </tr>
</table>

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_amazon
amazon = load_amazon()
print(amazon)
```

```python
# Forest fires in Rio and Sao Paulo
amazon.filter("state IN ('Rio', 'Sao Paulo')", print_info = False)
amazon["number"].plot(ts = "date", by = "state")
```
