# vDataFrame.scatter

```python
vDataFrame.scatter(columns: list,
                   catcol: str = "",
                   max_cardinality: int = 3,
                   cat_priority: list = [],
                   with_others: bool = True,
                   max_nb_points: int = 20000,
                   bbox: list = [],
                   img: str = "")
```

Draws a scatter plot of the input vcolumns.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td><div class="param_name">columns</div></td> <td><div class="type">list</div></td> <td><div class = "no">&#10060;</div></td> <td>List of the vcolumns names. The list must have two or three elements.</td> </tr>
    <tr> <td><div class="param_name">catcol</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Categorical vcolumn used to label the data.</td> </tr>
    <tr> <td><div class="param_name">max_cardinality</div></td> <td><div class="type">int</div></td> <td><div class = "yes">&#10003;</div></td> <td>Maximum number of distinct elements for 'catcol' to be used as categorical. The less frequent elements will be gathered together to create a new category: 'Others'.</td> </tr>
    <tr> <td><div class="param_name">cat_priority</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>List of the different categories to consider when labeling the data using the vcolumn 'catcol'. The other categories will be filtered.</td> </tr>
    <tr> <td><div class="param_name">with_others</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to false and the cardinality of the vcolumn 'catcol' is too big then the less frequent element will not be merged to another category and they will not be drawn.</td> </tr>
    <tr> <td><div class="param_name">max_nb_points</div></td> <td><div class="type">int</div></td> <td><div class = "yes">&#10003;</div></td> <td>Maximum number of points to display.</td> </tr>
    <tr> <td><div class="param_name">bbox</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>List of 4 elements to delimit the boundaries of the final Plot. It must be similar the following list: [xmin, xmax, ymin, ymax]</td> </tr>
    <tr> <td><div class="param_name">img</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Path to the image to display as background.</td> </tr>
</table>

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_iris
iris = load_iris()
print(iris)
```

```python
iris.scatter(["SepalWidthCm", "SepalLengthCm"])
```

```python
# 2D Scatter Plot with catcol
iris.scatter(["PetalWidthCm", "PetalLengthCm"], 
                 catcol = "Species")
```

```python
# 3D Scatter Plot
iris.scatter(["PetalWidthCm", "PetalLengthCm", "SepalLengthCm"], 
                 catcol = "Species")
```

```python
# Iris setosa vs others
iris.scatter(["PetalWidthCm", "PetalLengthCm"], 
                 catcol = "Species",
                 cat_priority = ["Iris-setosa"])
```

```python
# Only Iris setosa
iris.scatter(["PetalWidthCm", "PetalLengthCm"], 
                 catcol = "Species",
                 cat_priority = ["Iris-setosa"],
                 with_others = False)
```

```python
# img parameter can be used to display a background image
from vertica_ml_python import *
# Africa Dataset
africa = vDataFrame("africa_education")
africa = africa.groupby(["country_long", "lat", "lon"])
# displaying schools in Africa
africa.scatter(["lon", "lat"],
               catcol = "country_long",
               max_cardinality = 100,
               img = "img/africa.png")
```
