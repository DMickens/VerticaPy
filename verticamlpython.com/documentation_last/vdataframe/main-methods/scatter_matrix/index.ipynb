# vDataFrame.scatter_matrix

```python
vDataFrame.scatter_matrix(columns: list = [])
```

Draws the Scatter Matrix of the vDataFrame.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td><div class="param_name">columns</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>List of the vcolumns names. If empty, all the numerical vcolumns will be used.</td> </tr>
</table>

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_iris
iris = load_iris()
print(iris)
```

```python
iris.scatter_matrix()
```
