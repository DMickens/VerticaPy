# vDataFrame.testdf

```python
vDataFrame.testdf(ts: str, 
                  column: str, 
                  p: int = 1, 
                  alpha: float = 0.05,
                  by: list = [],
                  print_info: bool = True)
```

Dickey Fuller test (Time Series stationarity).

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td><div class="param_name">ts</div></td> <td><div class="type">str</div></td> <td><div class = "no">&#10060;</div></td> <td>vcolumn used as timeline. It will be used to order the data. It can be a numerical or type date like (date, datetime, timestamp...) vcolumn.</td> </tr>
    <tr> <td><div class="param_name">column</div></td> <td><div class="type">str</div></td> <td><div class = "no">&#10060;</div></td> <td>Input vcolumn to test.</td> </tr>
    <tr> <td><div class="param_name">p</div></td> <td><div class="type">int</div></td> <td><div class = "yes">&#10003;</div></td> <td>Number of lags to consider in the test.</td> </tr>
    <tr> <td><div class="param_name">alpha</div></td> <td><div class="type">float</div></td> <td><div class = "yes">&#10003;</div></td> <td>Significance Level. Probability to accept H0.</td> </tr>
    <tr> <td><div class="param_name">by</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>vcolumns used in the partition.</td> </tr>
    <tr> <td><div class="param_name">print_info</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to True, displays all the test information.</td> </tr>
</table>

### Returns

<b>tuple</b> : <br> [0] result of the test : True if H0 was rejected, <br>
               [1] gamma + 1          : gamma is the the Linear Regression lag1 coefficient., <br>
               [2] p_value            : p value of the coefficient.

### Example

```python
from vertica_ml_python.learn.datasets import load_smart_meters
sm = load_smart_meters()
sm["val"].plot(ts = "time", by = "id")
```

```python
sm.testdf(ts = "time",
          column = "val",
          p = 3,
          alpha = 0.01,
          by = ["id"])
```
