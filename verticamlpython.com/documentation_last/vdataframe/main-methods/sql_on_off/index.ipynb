# vDataFrame.sql_on_off

```python
vDataFrame.sql_on_off()
```

Displays each query generated by the vDataFrame.

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_titanic
titanic = load_titanic()
print(titanic)
```

```python
# Turning the SQL on
titanic.sql_on_off()
# Describing the Dataset
titanic.describe()
```
