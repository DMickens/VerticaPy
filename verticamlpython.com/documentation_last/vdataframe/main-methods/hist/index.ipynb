# vDataFrame.hist

```python
vDataFrame.hist(columns: list,
                method: str = "density",
                of: str = "",
                max_cardinality: tuple = (6, 6),
                h: tuple = (None, None),
                hist_type: str = "auto")
```

Draws the Histogram of the input vcolumns based on an aggregation.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td>columns</td> <td><div class="type">list</div></td> <td><div class = "no">&#10060;</div></td> <td>List of the vcolumns names. The list must have less than 5 elements.</td> </tr>
    <tr> <td><div class="param_name">method</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The method used to aggregate the data.<br>
                                                    <ul>
                                                        <li><b>count :</b> Number of elements.</li>
                                                        <li><b>density :</b> Percentage of the distribution.</li>
                                                        <li><b>mean :</b> Average of the vcolumn 'of'.</li>
                                                        <li><b>min :</b> Minimum of the vcolumn 'of'.</li>
                                                        <li><b>max :</b> Maximum of the vcolumn 'of'.</li>
                                                        <li><b>sum :</b> Sum of the vcolumn 'of'.</li>
                                                        <li><b>q% :</b> q Quantile of the vcolumn 'of' (ex: 50% to get the median).</li>
                                                        </ul></td> </tr>
    <tr> <td><div class="param_name">of</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The vcolumn used to compute the aggregation.</td> </tr>
    <tr> <td><div class="param_name">h</div></td> <td><div class="type">tuple</div></td> <td><div class = "yes">&#10003;</div></td> <td>Interval width of the vcolumns 1 and 2 bars. It is only valid if the vcolumns are numerical. Optimized h will be computed if the parameter is empty or invalid.</td> </tr>
    <tr> <td><div class="param_name">max_cardinality</div></td> <td><div class="type">tuple</div></td> <td><div class = "yes">&#10003;</div></td> <td>Maximum number of distinct elements for vcolumns 1 and 2 to be used as categorical (No h will be picked or computed)</td> </tr>
    <tr> <td><div class="param_name">hist_type</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The Histogram Type.<br>
                                                    <ul>
                                                        <li><b>auto :</b> Regular Histogram based on 1 or 2 vcolumns.</li>
                                                        <li><b>multi :</b> Multiple Regular Histograms based on 1 to 5 vcolumns.</li>
                                                        <li><b>stacked :</b> Stacked Histogram based on 2 vcolumns.</li></ul></td> </tr>
</table>

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_titanic
titanic = load_titanic()
print(titanic)
```

```python
titanic.hist(["survived", "pclass"])
```

```python
# Avg of survived partitioned by pclass, sex
titanic.hist(["sex", "pclass"], method = "avg", of = "survived")
```

```python
# Stacked : Avg of survived partitioned by pclass, sex
titanic.hist(["sex", "pclass"], 
            method = "avg", 
            of = "survived",
            hist_type = "stacked")
```

```python
# Multi Hist
titanic.hist(["age", "fare"], 
             method = "density", 
             hist_type = "multi")
```

```python
# Avg of survived partitioned by age, fare
# age will be discretized using an interval of 30 years
# fare will be discretized using an interval of 150£
titanic.hist(["age", "fare"], 
             method = "avg", 
             of = "survived",
             h = (30, 150))
```
