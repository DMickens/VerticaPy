# vDataFrame.hexbin

```python
vDataFrame.hexbin(columns: list,
                  method: str = "count",
                  of: str = "",
                  cmap: str = '',
                  gridsize: int = 20,
                  color: str = "white",
                  bbox: list = [],
                  img: str = "")
```

Draws the Hexbin of the input vcolumns based on an aggregation.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td>columns</td> <td><div class="type">list</div></td> <td><div class = "no">&#10060;</div></td> <td>List of the vcolumns names. The list must have one or two elements.</td> </tr>
    <tr> <td><div class="param_name">method</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The method used to aggregate the data.<br>
                                                    <ul>
                                                        <li><b>count :</b> Number of elements.</li>
                                                        <li><b>density :</b> Percentage of the distribution.</li>
                                                        <li><b>mean :</b> Average of the vcolumn 'of'.</li>
                                                        <li><b>min :</b> Minimum of the vcolumn 'of'.</li>
                                                        <li><b>max :</b> Maximum of the vcolumn 'of'.</li>
                                                        <li><b>sum :</b> Sum of the vcolumn 'of'.</li>
                                                        </ul></td> </tr>
    <tr> <td><div class="param_name">of</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The vcolumn used to compute the aggregation.</td> </tr>
    <tr> <td><div class="param_name">cmap</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Color Map.</td> </tr>
    <tr> <td><div class="param_name">gridsize</div></td> <td><div class="type">int</div></td> <td><div class = "yes">&#10003;</div></td> <td>Hexbin grid size.</td> </tr>
    <tr> <td><div class="param_name">color</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Color of the Hexbin borders.</td> </tr>
    <tr> <td><div class="param_name">bbox</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>List of 4 elements to delimit the boundaries of the final Plot. It must be similar the following list: [xmin, xmax, ymin, ymax]</td> </tr>
    <tr> <td><div class="param_name">img</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Path to the image to display as background.</td> </tr>
</table>

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_titanic
titanic = load_titanic()
print(titanic)
```

```python
titanic.hexbin(["age", "fare"])
```

```python
# Avg of survived partitioned by fare, age
titanic.hexbin(["age", "fare"], method = "avg", of = "survived")
```

```python
# img parameter can be used to display a background image
from vertica_ml_python import *
# Africa Dataset
africa = vDataFrame("africa_education")
# displaying average students score in Africa schools
africa.hexbin(["lon", "lat"],
              method = "avg",
              of = "zralocp",
              img = "img/africa.png")
```
