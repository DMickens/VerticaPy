# vDataFrame.autocorr

```python
vDataFrame.autocorr(ts: str, 
                    column: str, 
                    by: list = [],
                    p = 12,
                    method: str = "pearson", 
                    cmap: str = "",
                    round_nb: int = 3, 
                    focus: bool = False,
                    show: bool = True,
                    matrix: bool = False)
```

Computes the correlations of the input vcolumn and its lags.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
<tr> <td><div class="param_name">ts</div></td> <td><div class="type">str</div></td> <td><div class = "no">&#10060;</div></td> <td>TS (Time Series) vcolumn used to order the data. It can be of type date or a numerical vcolumn.</td> </tr>
<tr> <td><div class="param_name">column</div></td> <td><div class="type">str</div></td> <td><div class = "no">&#10060;</div></td> <td>Input vcolumn used to compute the Auto Correlation Plot.</td> </tr>
<tr> <td><div class="param_name">by</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>vcolumns used in the partition.</td> </tr>
<tr> <td><div class="param_name">p</div></td> <td><div class="type">int / list</div></td> <td><div class = "yes">&#10003;</div></td> <td>Int equals to the maximum number of lag to consider during the computation or List of the different lags to include during the computation. p must be positive or a list of positive integers.</td> </tr>
    <tr> <td><div class="param_name">unit</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Unit used to compute the lags.<br>
                                                    <ul>
                                                        <li><b>rows :</b> Natural lags.</li>
                                                        <li><b>else :</b> Any time unit, for example you can write 'hour' to compute the hours lags or 'day' to compute the days lags.</li>
                                                        </ul></td> </tr>
    <tr> <td><div class="param_name">method</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Method used to compute the correlation.<br>
                                                    <ul>
                                                        <li><b>pearson :</b> Pearson correlation coefficient (linear).</li>
                                                        <li><b>spearmann :</b> Spearmann correlation coefficient (monotonic - rank based).</li>
                                                        <li><b>kendall :</b> Kendall correlation coefficient (similar trends).</li>
                                                        <li><b>cramer :</b> Cramer's V (correlation between categories).</li>
                                                        <li><b>biserial :</b> Biserial Point (correlation between binaries and a numericals).</li>
                                                        </ul></td> </tr>
    <tr> <td><div class="param_name">cmap</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Color Map.</td> </tr>
    <tr> <td><div class="param_name">round_nb</div></td> <td><div class="type">int</div></td> <td><div class = "yes">&#10003;</div></td> <td>Round the coefficient using the input number of digits. It is used only if parameter 'matrix' is set to True.</td> </tr>
    <tr> <td><div class="param_name">focus</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>Computes a correlation vector instead of the matrix. It is used only if parameter 'matrix' is set to True.</td> </tr>
    <tr> <td><div class="param_name">show</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to True, the Auto Correlation Plot will be drawn using Matplotlib.</td> </tr>
    <tr> <td><div class="param_name">matrix</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to True, draw a correlation matrix instead of a regular plot.</td> </tr>
</table>

### Returns

<a href="../../../utilities/tablesample">tablesample</a> : An object containing the result. For more information, check out <a href="../../../utilities/tablesample">utilities.tablesample</a>.

### Example

```python
from vertica_ml_python.learn.datasets import load_amazon
amazon = load_amazon()
print(amazon)
```

```python
# Autocorrelation Plot for each 'year' lag using Spearman coefficients
# p = 48: it will compute 48 'months' lags
amazon.autocorr(ts = "date", 
                column = "number", 
                p = 48,
                by = ["state"],
                unit = "month",
                method = "spearman")
```

```python
# Autocorrelation Plot using only the selected lags
amazon.autocorr(ts = "date", 
                column = "number", 
                by = ["state"],
                p = [1, 3, 6, 7],
                unit = "year",
                method = "pearson")
```

```python
# Autocorrelation Matrix for each 'month' lag
amazon.autocorr(ts = "date", 
                column = "number", 
                by = ["state"],
                p = 12,
                unit = "month",
                method = "pearson", 
                round_nb = 3, 
                matrix = True)
```

```python
# Autocorrelation Vector for each 'year' lag
amazon.autocorr(ts = "date", 
                column = "number", 
                by = ["state"],
                p = 12,
                unit = "year",
                method = "pearson", 
                round_nb = 3, 
                focus = True,
                matrix = True)
```
