# vDataFrame.cov

```python
vDataFrame.cov(columns: list = [], 
               cmap: str = "",
               focus: str = "",
               show: bool = True)
```

Computes the Covariance Matrix of the vDataFrame. 

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
<tr> <td><div class="param_name">columns</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>List of the vcolumns names. If empty, all the numerical vcolumns will be used.</td> </tr>
    <tr> <td><div class="param_name">cmap</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Color Map.</td> </tr>
    <tr> <td><div class="param_name">focus</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Focus the computation on only one vcolumn.</td> </tr>
    <tr> <td><div class="param_name">show</div></td> <td><div class="type">bool</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to True, the Covariance Matrix will be drawn using Matplotlib..</td> </tr>
</table>

### Returns

<a href="../../../utilities/tablesample">tablesample</a> : An object containing the result. For more information, check out <a href="../../../utilities/tablesample">utilities.tablesample</a>.

### Example

```python
from vertica_ml_python.learn.datasets import load_titanic
titanic = load_titanic()
print(titanic)
```

```python
titanic.cov()
```

```python
titanic.cov(focus = "survived")
```
