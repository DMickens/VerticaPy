### Example

```python
# SVM
from vertica_ml_python.learn.svm import LinearSVC
model = LinearSVC(name = "public.SVC_iris")
model.fit("public.iris_clean", 
          ["PetalLengthCm", "PetalWidthCm"], 
          "Species_Iris-setosa")
model.plot()
```

```python
# Logit
from vertica_ml_python.learn.linear_model import LogisticRegression
model = LogisticRegression(name = "public.SVC_titanic")
model.fit("public.titanic", 
          ["fare", "age"], 
          "survived")
model.plot()
```
