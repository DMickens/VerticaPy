# unsupervised.plot

```python
unsupervised.plot(X: list = [], 
                  tablesample: float = -1)
```

Draws the model if the number of predictors is 2 or 3.

### Parameters

<table id="parameters">
    <tr> <td><div class="param_name">X</div></td> <td><div class="type">list</div></td> <td><div class = "yes">&#10003;</div></td> <td>List of the predictors to display. The score will stay the same.</td> </tr>
    <tr> <td><div class="param_name">tablesample</div></td> <td><div class="type">float</div></td> <td><div class = "yes">&#10003;</div></td> <td>Sample of data to display. If this number is not between 0 and 1 all the data points will be displayed.</td> </tr>
</table>

### Example

```python
from vertica_ml_python.learn.neighbors import LocalOutlierFactor
model = LocalOutlierFactor(name = "public.LOF_iris")
model.fit("public.iris", ["PetalLengthCm", "SepalLengthCm"])
model.plot()
```
