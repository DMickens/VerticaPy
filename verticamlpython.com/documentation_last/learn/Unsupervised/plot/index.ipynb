# unsupervised.plot

```python
unsupervised.plot(voronoi: bool = False)
```

Draws the unsupervised model.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td><div class="param_name">input_relation</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>If set to True, a voronoi plot will be drawn. It is only available for KMeans using 2 predictors.</td> </tr>
</table>

### Example

```python
from vertica_ml_python.learn.cluster import KMeans
model = KMeans(name = "public.KMeans_iris",
               n_cluster = 3)
model.fit("public.iris", ["PetalLengthCm", "SepalLengthCm"])
# Regular Plot
model.plot()
```

```python
model = KMeans(name = "public.KMeans_iris",
               n_cluster = 8)
model.fit("public.iris", ["PetalLengthCm", "SepalLengthCm"])
# Voronoi plot
model.plot(voronoi = True)
```
