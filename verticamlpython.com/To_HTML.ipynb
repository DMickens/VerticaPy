# Converting Vertica ML Python PHP website into HTML

```python
# Import the OS file
import os

# Loading the main include elements
def get_includes(dir_nb: int = 0):
    # Header
    header = open("./include/header.php", "r")
    header = "".join(header.readlines())
    header = header.replace("http://localhost:666/verticamlpython.com/",
                            "../" * dir_nb).replace(".php", ".html")
    header = header.replace("http://localhost:666/verticamlpython.com",
                            "../" * dir_nb + "index.html").replace(".php", ".html")
    # Footer
    footer = open("./include/footer.php", "r")
    footer = "".join(footer.readlines())
    footer = footer.replace("http://localhost:666/verticamlpython.com/",
                            "../" * dir_nb).replace(".php", ".html")
    # Head
    head = open("./include/head.php", "r")
    head = "".join(head.readlines())
    head = head.replace("http://localhost:666/verticamlpython.com/",
                        "../" * dir_nb).replace(".php", ".html")
    # Logo
    logo = open("./include/logo.php", "r")
    logo = "".join(logo.readlines())
    return (head, logo, header, footer)

def replace_vml_index(path: str, dir_nb: int):
    head, logo, header, footer = get_includes(dir_nb)
    for filename in os.listdir(path):
        if os.path.isdir(path + "/" + filename) and (filename not in ('Classification', 
                                                                      'Regression', 
                                                                      'Unsupervised',
                                                                      'datasets',
                                                                      'examples',
                                                                      'metrics',
                                                                      'img')):
            index_include = open(path + "/" + filename + '/index-include.php', "r")
            index_include = "".join(index_include.readlines())
            index = open(path + "/" + filename + '/index.php', "r")
            index = "".join(index.readlines())
            index = index.replace("<?php include('" + "../" * dir_nb + "include/head.php'); ?>", head)
            index = index.replace("<?php include('" + "../" * dir_nb + "include/header.php'); ?>", header)
            index = index.replace("<?php include('" + "../" * dir_nb + "include/logo.php'); ?>", logo)
            index = index.replace("<?php include('" + "../" * dir_nb + "include/footer.php'); ?>", footer)
            index = index.replace("<?php include('index-include.php'); ?>", index_include)
            index = index.replace('.php', '.html')
            index_html = open(path + "/" + filename + '/index.html', "w+")
            index_html.write(index)
            os.remove(path + "/" + filename + '/index-include.php')
            os.remove(path + "/" + filename + '/index.php')
            try:
                os.remove(path + "/" + filename + '/index.ipynb')
            except:
                pass

def replace_vml_index_folder(path: str, dir_nb: int):
    head, logo, header, footer = get_includes(dir_nb)
    index_include = open(path + '/index-include.php', "r")
    index_include = "".join(index_include.readlines())
    index = open(path + '/index.php', "r")
    index = "".join(index.readlines())
    index = index.replace("<?php include('" + "../" * dir_nb + "include/head.php'); ?>", head)
    index = index.replace("<?php include('" + "../" * dir_nb + "include/header.php'); ?>", header)
    index = index.replace("<?php include('" + "../" * dir_nb + "include/logo.php'); ?>", logo)
    index = index.replace("<?php include('" + "../" * dir_nb + "include/footer.php'); ?>", footer)
    index = index.replace("<?php include('index-include.php'); ?>", index_include)
    index = index.replace('.php', '.html')
    index_html = open(path + '/index.html', "w+")
    index_html.write(index)
    os.remove(path + '/index-include.php')
    os.remove(path + '/index.php')
    try:
        os.remove(path + '/index.ipynb')
    except:
        pass

def replace_vml_php(path: str, dir_nb: int):
    head, logo, header, footer = get_includes(dir_nb)
    for filename in os.listdir(path):
        if filename.endswith(".php") and filename not in ('quick-start-include.php',
                                                          'connection-include.php'):
            file = open(path + "/" + filename, "r")
            file = "".join(file.readlines())
            file = file.replace("<?php include('" + "../" * dir_nb + "include/head.php'); ?>", head)
            file = file.replace("<?php include('" + "../" * dir_nb + "include/header.php'); ?>", header)
            file = file.replace("<?php include('" + "../" * dir_nb + "include/logo.php'); ?>", logo)
            file = file.replace("<?php include('" + "../" * dir_nb + "include/footer.php'); ?>", footer)
            if filename == 'installation.php':
                file_tmp = open("connection-include.php", "r")
                file_tmp = "".join(file_tmp.readlines())
                file = file.replace("<?php include('connection-include.php'); ?>", file_tmp)
            elif filename == 'quick-start.php':
                file_tmp = open("quick-start-include.php", "r")
                file_tmp = "".join(file_tmp.readlines())
                file = file.replace("<?php include('quick-start-include.php'); ?>", file_tmp)
            file = file.replace('.php', '.html')
            file_html = open(path + "/" + filename.replace('.php', '.html'), "w+")
            file_html.write(file)
            os.remove(path + "/" + filename)
            if filename == 'installation.php':
                os.remove("connection-include.php")
            elif filename == 'quick-start.php':
                os.remove("quick-start-include.php")

replace_vml_index("./documentation_last/vdataframe/main-methods", 4)
replace_vml_index("./documentation_last/vdataframe/vcolumn-methods", 4)
replace_vml_php("./documentation_last/vdataframe", 2)
replace_vml_index_folder("./documentation_last/vdataframe/object", 3)

replace_vml_index("./documentation_last/utilities", 3)
replace_vml_php("./documentation_last/utilities", 2)

replace_vml_index("./documentation_last/connect", 3)
replace_vml_php("./documentation_last/connect", 2)

replace_vml_php("./documentation_last", 1)
replace_vml_index("./documentation_last/learn", 3)
replace_vml_php("./documentation_last/learn", 2)
replace_vml_index("./documentation_last/learn/Classification", 4)
replace_vml_index("./documentation_last/learn/Regression", 4)
replace_vml_index("./documentation_last/learn/Unsupervised", 4)
replace_vml_index("./documentation_last/learn/datasets", 4)
replace_vml_php("./documentation_last/learn/datasets", 3)
replace_vml_index("./documentation_last/learn/examples", 4)
replace_vml_php("./documentation_last/learn/examples", 3)
replace_vml_index("./documentation_last/learn/metrics", 4)
replace_vml_php("./documentation_last/learn/metrics", 3)

replace_vml_php("./workshop", 1)
replace_vml_index("./workshop/data_exploration", 3)
replace_vml_index("./workshop/data_prep", 3)
replace_vml_index("./workshop/introduction", 3)
replace_vml_index("./workshop/ml", 3)
replace_vml_index("./workshop/full_stack", 3)
replace_vml_index_folder("./workshop/data_ingestion", 2)
replace_vml_php("./workshop/data_exploration", 2)
replace_vml_php("./workshop/data_prep", 2)
replace_vml_php("./workshop/introduction", 2)
replace_vml_php("./workshop/ml", 2)
replace_vml_php("./workshop/full_stack", 2)

replace_vml_index("./examples", 2)
replace_vml_php("./examples", 1)

replace_vml_php("./", 0)

import shutil
shutil.rmtree('include')
shutil.rmtree('./img/examples/high')
os.remove('quick-start.ipynb')
os.remove('connection.ipynb')
```
